# Data Visualization with Seaborn
This notebook demonstrates various Seaborn plots for exploratory data analysis.

```python

# Importing Required Libraries
import seaborn as sns
import numpy as np
import pandas as pd


# Create Sample Dataset
data = pd.DataFrame({
    "total_bill": np.random.uniform(10, 50, 200),
    "tip": np.random.uniform(2, 10, 200),
    "sex": np.random.choice(["Male", "Female"], size=200),
    "day": np.random.choice(["Thur", "Fri", "Sat", "Sun"], size=200),
    "time": np.random.choice(["Lunch", "Dinner"], size=200),
    "size": np.random.randint(1, 6, size=200)
})

# Scatter Plot
sns.scatterplot(data=data, x="total_bill", y="tip", hue="sex", style="time", size="size")

# Histogram
sns.histplot(data=data, x="total_bill", bins=20, kde=True)

# Bar Plot
sns.barplot(data=data, x="day", y="total_bill", hue="sex")

# Box Plot
sns.boxplot(data=data, x="day", y="total_bill", hue="sex")

# Violin Plot
sns.violinplot(data=data, x="day", y="total_bill", hue="sex", split=True)

# Strip Plot
sns.stripplot(data=data, x="day", y="total_bill", hue="sex", dodge=True)

# Pair Plot
sns.pairplot(data=data, hue="sex")

# Heatmap
sns.heatmap(data.select_dtypes(include="number").corr(), annot=True, cmap="coolwarm", fmt=".2f")

# Facet Grid
sns.FacetGrid(data, col="sex", row="time").map(sns.scatterplot, "total_bill", "tip")

# Regression Plot
sns.lmplot(data=data, x="total_bill", y="tip", hue="sex")

# Changing Style
sns.set_style("whitegrid")

# Count Plot (Shows counts of categorical variable levels)
sns.countplot(data=data, x="day", hue="sex")

# Joint Plot (Displays the relationship between two variables with additional distributions)
sns.jointplot(data=data, x="total_bill", y="tip", kind="hex")

# KDE Plot (Kernel Density Estimation plot, a smooth version of histogram)
sns.kdeplot(data=data, x="total_bill", hue="sex", fill=True)

# Swarm Plot (A categorical scatter plot to show all data points, avoiding overlap)
sns.swarmplot(data=data, x="day", y="total_bill", hue="sex", dodge=True)

# PairGrid (Grid-based visualization for pairwise relationships)
g = sns.PairGrid(data, hue="sex")
g.map_lower(sns.kdeplot, cmap="Blues")
g.map_diag(sns.histplot, kde_kws={'color': 'black'})

# Displot (Distribution plot with flexibility in histogram and KDE)
sns.displot(data=data, x="total_bill", kde=True, hue="sex", bins=15)

# Regression with Confidence Interval (lmplot with the confidence interval disabled)
sns.lmplot(data=data, x="total_bill", y="tip", hue="sex", ci=None)

# Aspect Ratio Adjustment (Adjust figure size with sns.set_context)
sns.set_context("talk")  # Adjusts the context for larger text and better readability
```
