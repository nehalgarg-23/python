```python

import matplotlib.pyplot as plt

# Basic line plot
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]
plt.plot(x, y, marker='o', color='blue', label='Prime numbers')
plt.title("Basic Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.grid(True)
plt.show()
```

```python

import matplotlib.pyplot as plt

# Bar plot
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 8, 5]
plt.bar(categories, values, color='purple', alpha=0.7)
plt.title("Bar Plot Example")
plt.xlabel("Categories")
plt.ylabel("Values")
plt.show()
```

```python

import matplotlib.pyplot as plt

# Scatter plot
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 3, 5]
plt.scatter(x, y, color='red', label='Data points')
plt.title("Scatter Plot Example")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.legend()
plt.show()
```

```python

import matplotlib.pyplot as plt

# Pie chart
sizes = [15, 30, 45, 10]
labels = ['A', 'B', 'C', 'D']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0, 0.1, 0, 0)  # explode the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title("Pie Chart Example")
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Histogram
data = np.random.randn(1000)
plt.hist(data, bins=20, color='blue', edgecolor='black', alpha=0.7)
plt.title("Histogram Example")
plt.xlabel("Bins")
plt.ylabel("Frequency")
plt.show()
```

```python

import matplotlib.pyplot as plt
import numpy as np

# Subplots
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(x, y1, label='sin(x)', color='green')
plt.title("Sine Function")
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(x, y2, label='cos(x)', color='orange')
plt.title("Cosine Function")
plt.legend()

plt.tight_layout()
plt.show()
```
